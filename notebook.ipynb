```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
from IPython.display import Audio, display
import pedalboard
import effects.modulation as ModulationFX
import effects.time_based as TimeBasedFX
import effects.quality_reduction as QualityReductionFX

# Load audio file
DURATION_IN_SECONDS = 5
OFFSET_IN_SECONDS = 0
inspector_norse = "/Users/ellie/Music/Tracks/All/Inspector Norse.mp3"
sine_wave = "/Users/ellie/Downloads/400 Hz Test Tone.mp3"
original_audio, sr = librosa.load(path=sine_wave, offset=OFFSET_IN_SECONDS, duration=DURATION_IN_SECONDS)

# Create pedalboard with desired effects and process audio
board = pedalboard.Pedalboard([
    ModulationFX.phaser,
    # ModulationFX.chorus,
    # TimeBasedFX.reverb,
    # TimeBasedFX.delay,
    # QualityReductionFX.bitcrush,
    # QualityReductionFX.resample,
])
manipulated_audio = board.process(original_audio, sr)

def display_waveform(original_audio, manipulated_audio, sr):
    plt.figure(figsize=(30, 5))
    plt.title("Original (blue) vs. Manipulated Waveform (red)")
    librosa.display.waveshow(original_audio, sr=sr, color="blue")
    librosa.display.waveshow(manipulated_audio, sr=sr, color="red")
display_waveform(original_audio, manipulated_audio, sr)

def display_audio_players(original_audio, manipulated_audio, sr):
    print("original audio:")
    display(Audio(original_audio, rate=sr))
    print("manipulated audio:")
    display(Audio(manipulated_audio, rate=sr))
display_audio_players(original_audio, manipulated_audio, sr)

```
